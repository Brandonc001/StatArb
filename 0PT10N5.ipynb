```python
data = yf.download(tickers="AAPL SPY TSLA MSFT", start='2020-01-01')['Adj Close']
#yf.option_chain('AAPL',date='2022-12-23').calls[{'strike', 'bid', 'ask', 'volume'}]
#spy = yf.download('spy', start='2020-01-01')
spy = yf.Ticker("spy")
tsla = yf.Ticker("tsla")
aapl = yf.Ticker("aapl")
msft = yf.Ticker("msft")
adbe = yf.Ticker("adbe")
amd = yf.Ticker("amd")
goog = yf.Ticker("goog")
amzn = yf.Ticker("amzn")
amgn = yf.Ticker("amgn")
qqq = yf.Ticker("qqq")
ticks = [spy, tsla, aapl, msft, adbe, amd, goog, amzn, amgn, qqq]
```

```python
### strike - stock + call - put
```

```python
sets = {}
for x in ticks:
    stock = x.history(start="2022-06-02", interval="1d")['Close'][-1]
    call = x.option_chain(date='2022-12-23').calls[{'strike', 'bid', 'ask', 'volume'}]
    call['mid_x'] = (call['bid']+call['ask'])/2
    put = x.option_chain(date='2022-12-23').puts[{'strike', 'bid', 'ask', 'volume'}]
    put['mid_y'] = (put['bid']+put['ask'])/2
    opt =  pd.merge(put, call, left_on='strike', right_on='strike')
    opt['par1'] = opt['mid_x'] + stock
    opt['par2'] = opt['mid_y']+ opt['strike']
    opt['par'] = opt['strike']-stock+opt['mid_x']-opt['mid_y']
    opt = opt[opt['volume_x']>100.0]
    opt = opt[opt['volume_y']>100.0]
    opt = opt.sort_values('par', ascending=False)
    sets[str(x)] = opt[{'volume_x', 'volume_y', 'strike', 'par'}]
```

```python
result = pd.concat(sets)
result
```
